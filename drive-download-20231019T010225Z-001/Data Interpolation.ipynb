Data Interpolation : Data interpolation is the process of estimating unknown values within a dataset based on the known values. In python, there are various liabraries
available that can be used for data interpolation, such as numpy, Scipy and pandas. here is an example of how to perform data interpolationusing the numoy library.

```python
1. Linear interpolation
```

```python
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
Cubic interpolation
```

```python
from scipy.interpolate import interp1d
import numpy as np
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 8, 27, 32, 125])

f = interp1d(x, y, kind = 'cubic')
```

```python
x_new = np.linspace(1, 5, 10)
y_interp = f(x_new)
print(y_interp)
```

```python
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([1,4,9, 16,25])
```

```python
p = np.polyfit(x,y,2)
```

```python
x_new = np.linspace(1,5,10)
y_interp = np.polyval(p, x_new)
```

```python
plt.scatter(x, y)
```
