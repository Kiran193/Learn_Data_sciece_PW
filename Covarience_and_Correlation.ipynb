```python
import seaborn as sns
df = sns.load_dataset('healthexp')
df.head()
```

```python
## Covareince
import numpy as np

# Extract the numerical columns from the DataFrame
numerical_data = df.select_dtypes(include='number')

# Calculate the covariance matrix
covariance_matrix = np.cov(numerical_data, rowvar=False)

covariance_matrix
```

```python
df.cov()
```

```python
#pearson correlation
df.corr(method='pearson')
```

```python
#spearman rank correlation
df.corr(method='spearman')
```

```python
import pandas as pd

# Create a sample DataFrame
data = {'Column1': [1, 2]}
df = pd.DataFrame(data)

# Calculate the covariance for the same column
covariance = df['Column1'].cov(df['Column1'])

print("Covariance between Column1 and itself:", covariance)

# (1-1.5)*(1-1.5)+(2-1.5)(2-1.5)/2-1 ===> 0.5
# 2-1 == n-1
```
